```python
regions_list = list(regions.Group.unique())
regions_list
```

```python
 # Turn population into bubble sizes. Use min_size and factor to tweak.
scale_factor = 200
population_size = np.sqrt(population / np.pi) / scale_factor
population_size.head()
```

```python
min_size=3
population_size.where(population_size >= min_size).head()
```

```python
population_size = population_size.where(population_size >= min_size)\
.fillna(min_size)
population_size.head()
```
