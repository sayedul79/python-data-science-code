### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Read dataset

```python
students=pd.read_csv("data/StudentsPerformance.csv")
students.head()
```

```python
students.columns
```

## Using the distribution plot

```python
sns.displot(x="math score", data=students)#kind="hist"
```

```python
sns.displot(x="math score", data=students, 
           kind="hist", hue="gender")
```

```python
sns.displot(x="math score", data=students, 
           kind="hist", hue="gender", 
           col="gender")
```

```python
hue_colors=['g', 'orange']
sns.displot(x="math score", data=students, 
           kind="hist", hue="gender", 
           col="gender", palette=hue_colors)
```

### Adding kde

```python
sns.displot(x="math score", data=students, kde=True)
```

### Adding rug

```python
sns.displot(x="math score", data=students, kde=True, rug=True)
```

### Kernel density estimation (kde)

```python
sns.displot(x="math score", data=students, kind="kde")
```

```python
sns.displot(y="math score", data=students, kind="kde")
```

### Add hue

```python
sns.displot(x="math score", data=students, kind="kde", hue="gender")
```

### Fill the plot with color

```python
sns.displot(x="math score", data=students, kind="kde",
           hue="gender", fill=True)
```

```python
sns.displot(x="math score", data=students, kind="kde",
           hue="gender",
           fill=True, alpha=0.5, 
           linewidth=0)
```

```python
sns.displot(x="math score", data=students, kind="kde",
           hue="gender", multiple="stack")
```

```python
sns.displot(x="math score", data=students, kind="kde", rug=True)
```

### Empirical distribution function (ecdf)

```python
sns.displot(x="math score", data=students, kind="ecdf")
```

### Draw a bivariate plot, assign both x and y

```python
sns.displot(x="reading score", y="math score", data=students)
```

```python
sns.displot(x="reading score", y="math score", data=students,
           kind="kde")
```

```python
sns.displot(x="reading score", y="math score", data=students,
           kind="kde", rug=True)
```

## Regression Plots in Seaborn

```python
sns.regplot(x="reading score", y="writing score", data=students)
```

### `regplot()`

```python
sns.regplot(x="reading score", y="writing score", data=students)
#There are no option for hue
```

### `lmplot()`

```python
sns.lmplot(x="reading score", y="writing score", data=students,
          hue="gender")
```

```python
sns.lmplot(x="reading score", y="writing score", data=students,
          hue="gender", row="gender")
```
