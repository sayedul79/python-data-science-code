```python
df=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
df.head()
```

```python
df.shape
```

```python
df[df["Country/Region"].isin(["Brazil"])]
```

```python
df_desire=df[df["Country/Region"].isin(["Bangladesh", "Pakistan", "India", 
                          "Iran", "US", "Italy", "Spain", "Brazil", 
                          "Uk", "China"])]
df_desire.head()
```

```python
df_desire_melt=pd.melt(df_desire, id_vars=["Province/State", 
                                          "Country/Region", 
                                          "Lat", 
                                          "Long"], 
                      var_name="date", 
                      value_name="total_cases")
df_desire_melt.head()
```

```python
df_desire_melt.drop(["Province/State", "Lat", "Long"], 
              inplace=True, 
              axis=1)
df_desire_melt.head()
```

```python
df_desire_melt["date"]=pd.to_datetime(df_desire_melt.date)
df_desire_melt.head()
```

```python
df_desire_pivot=pd.pivot_table(df_desire_melt, columns="Country/Region", 
              values="total_cases", 
              index="date", aggfunc="sum")
df_desire_pivot.head()
```

```python
bcr.bar_chart_race(df_desire_pivot.iloc[-100:, :])
```
