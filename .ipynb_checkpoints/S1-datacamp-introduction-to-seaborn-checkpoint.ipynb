```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Scatter plot

```python
height = [62, 64, 69, 75, 66,
            68, 65, 71, 76, 73]

weight = [120, 136, 148, 175, 137,
            165, 154, 172, 200, 187]
sns.scatterplot(x=height, y=weight)
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

### A scatter plot with hue

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

### Setting hue order

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", 
               hue_order=["No", "Yes"])
```

### Specifying hue colors

```python
hue_colors={'Yes':'m', 'No':'y'}
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker",
               palette=hue_colors)
```

```python
hue_colors = {"Yes": "#808080", "No": "#00FF00"}
sns.scatterplot(x="total_bill", y="tip", data=tips, 
                hue="smoker",palette=hue_colors)
```

### Introducing `relplot()`

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter")
```

### Subplots in columns

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter",
        col="smoker")
```

### Subplots in rows

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter",
        row="smoker")
```

### Subplots in rows and columns

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter",
        col="smoker",
        row="time")
```

### Wrapping columns

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter",
        col="day",
        col_wrap=2)
```

### Ordering columns

```python
sns.relplot(x="total_bill",
        y="tip",
        data=tips,
        kind="scatter",
        col="day",
        col_wrap=2,
        col_order=["Thur",
                    "Fri",
                    "Sat",
                    "Sun"])
```

### Hue and scatter plots

```python
student_data=pd.read_csv("data/datacamp/student-alcohol-consumption.csv")
student_data.columns
```

```python
sns.scatterplot(x='absences', y='G3', hue='location', data=student_data)
```

### Hue and count plots

```python
 #Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural': "green", 'Urban': "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x='school', hue='location', palette=palette_colors, 
              data=student_data)
```

### Creating subplots with col and row

```python
sns.relplot(x="absences", y="G3", 
                data=student_data, kind='scatter')
```

```python
#Change to make subplots based on study time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", col='study_time')
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")
```

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x='G1', y='G3', data=student_data, kind='scatter')
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", col='schoolsup', col_order=['yes', 'no'])
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"], row='famsup', row_order=['yes', 'no'])
```

### Changing the size of scatter plot points

```python
mpg=pd.read_csv("data/datacamp/mpg.csv")
mpg.columns
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x='horsepower', y='mpg', size='cylinders', data=mpg, 
            kind='scatter')
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders", hue='cylinders')
```

### Changing the hue of scatter plot point

```python
# Create scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration', y='mpg', data=mpg, 
            kind='scatter', hue='origin')
```

### Changing the style of scatter plot point

```python
# Create scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration', y='mpg', data=mpg, kind='scatter',
           hue='origin', style='origin')
```

### Interpreting line plots

```python
# Create line plot
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line')
```

### Visualizing standard deviation with line plots

```python
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')
```

```python
# Create line plot of model year vs. horsepower
sns.relplot(x="model_year", y="horsepower",
            data=mpg, kind="line", ci=None)
```

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, hue='origin', style='origin')
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True, dashes=False)
```
