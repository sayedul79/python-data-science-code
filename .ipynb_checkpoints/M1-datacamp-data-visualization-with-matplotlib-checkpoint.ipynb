### Import library

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Import dataset

```python
df1=pd.read_csv('data/datacamp/seattle_weather.csv')
df2=pd.read_csv('data/datacamp/austin_weather.csv')
```

```python
df1.iloc[:5, :5]
```

```python
df2.iloc[:5, :5]
```

```python
seattle_weather=df1.loc[:11, ["DATE","MLY-TAVG-NORMAL", "MLY-PRCP-NORMAL", 
                             "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL"]]
seattle_weather.rename(columns={'DATE': 'MONTH'}, inplace=True)
seattle_weather.tail()
```

```python
austin_weather=df2.loc[:11, ["DATE","MLY-TAVG-NORMAL", "MLY-PRCP-NORMAL", 
                            "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL"]]
austin_weather.rename(columns={'DATE': 'MONTH'}, inplace=True)
austin_weather.tail()
```

## Introduction to data visualization with Matplotlib

### Introducing the pyplot interface

```python
fig, ax = plt.subplots()
plt.show()
```

### Adding data to axes

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(austin_weather.MONTH, austin_weather["MLY-TAVG-NORMAL"])
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

## Customizing your plots

### Customizing dat appearance

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o')
```

### [Chossing marker](https://matplotlib.org/api/markers_api.html)

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='v')
```

### Setting the linestyle

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o', linestyle='--')
```

### Eliminiting lines with linestyle

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o', linestyle='None')
```

### Choosing color

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o', linestyle='--', color='r')
```

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o', linestyle='--', color='r', markerfacecolor='b')
```

### Customizing the axes label

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],
       marker='o', linestyle='--', color='r')

ax.set_xlabel("Time (months)")

ax.set_ylabel("Average temperature (Fahrenheit degrees)")

ax.set_title("Weather in Seattle")
```

## Small multiples
### Adding data

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"],
    seattle_weather["MLY-PRCP-NORMAL"],
    color='b')
ax.set_xlabel("Time (months)")
ax.set_ylabel("Precipitation (inches)")
```

### Adding more data

```python
fig, ax=plt.subplots()
ax.plot(seattle_weather["MONTH"],seattle_weather["MLY-PRCP-NORMAL"], 
        linestyle="-", color='m')

ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"],
        linestyle='--', color='b')

ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
        linestyle='--', color='r')
```

### Small multiples with `plt.subplots`

```python
fig,ax=plt.subplots(3,2)
```

```python
ax.shape
```

```python
fig,ax=plt.subplots(3,2)
ax[0, 0].plot(seattle_weather["MONTH"],
        seattle_weather["MLY-PRCP-NORMAL"], color='b')
```

### Subplots with data

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 
           color='b')

ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"], 
           linestyle='--', color='b')

ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
           linestyle='--', color='b')


ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
           color='r')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"],
           linestyle='--', color='r')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"],
           linestyle='--', color='r')


ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
```

### Sharing the y axis range

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharey=True)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], 
           color='b')

ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"], 
           linestyle='--', color='b')

ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"],
           linestyle='--', color='b')


ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
           color='r')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"],
           linestyle='--', color='r')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"],
           linestyle='--', color='r')


ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
```
