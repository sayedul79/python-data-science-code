### Relationship analysis

```python
correlation=students.corr()
correlation
```

```python
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(students)
```

```python
sns.relplot(x='math score', y='reading score', hue='gender', 
            data=students)
```

```python
sns.relplot(x='math score', y='reading score', hue='lunch', 
            data=students)
```

```python
sns.distplot(a=students['reading score'])
```

```python
sns.distplot(a=students['math score'])
```

```python
sns.distplot(a=students['writing score'])
```

```python
sns.catplot(x=students['gender'], y=students['math score'], data=students)
```

```python
sns.catplot(x=students['gender'], y=students['math score'], data=students,
           kind='box')
```

```python
sns.catplot(x='gender', y='math score', data=students, kind='box')
```

```python
sns.catplot(x='gender', y='math score', data=students, kind='box',
           hue='lunch')
```
